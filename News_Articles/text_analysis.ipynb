```python
import os # import os to get all the names
```

```python
filename_array = os.listdir('.') # return all the news filenames
filename_array.pop(filename_array.index('text_analysis.ipynb')) # remove the jupyter notebook from namelist
filename_array.pop(filename_array.index('compiled_news.txt')) # remove the jupyter notebook from namelist
print(filename_array)
```

** below is to rename the news files **

```python
# iterate through news files and perform simple data cleaning
for filename in filename_array:
    # print(filename)
    current_file = open(filename, "r", encoding="utf-8", errors='ignore')
    current_content = '' # empty string to hold the text file content
    for line in current_file:
        line.replace('+','') # remove unrecognized string

        if 'Date Published to Web:' in line: # check date to name the new file
            date = line[23:-1]
            date = date.replace('/','-') # replace / to - for easier recognition
            if date.find('-') == 1:
                date = '0'+ date
            if date.find('-',4,6) ==4:
                date = date[0:3] + '0' + date[3:]

            # now that date is MM-DD-YYYY format, change to YYYY-MM-DD for time sequence
            new_date = date[-4:]+'-'+date[0:2] + '-' + date[-7:-5] # add original filename to distinquish news with same date
            new_filename = new_date + '-' + filename[:-4] +'.txt'
            line = 'Date Published to Web:' + new_date + '\n'

        current_content += line # add processed line to current_content

    # write current cleaned content into newly named file
    with open(new_filename, "w") as new_file:
        new_file.write(current_content)
    new_file.close()
    current_file.close()

    
```

Compile all news into one file to facilitate text analysis

```python
filename_array = os.listdir('.') # return all the news filenames
filename_array.pop(filename_array.index('text_analysis.ipynb')) # remove the jupyter notebook from namelist
filename_array.pop(filename_array.index('compiled_news.txt')) # remove the jupyter notebook from namelist
filename_array.sort()
print(filename_array)
```

```python
compiled_news = open('compiled_news.txt','w') 
count = 0
for filename in filename_array:
    count+=1
    current_file = open(filename, "r", encoding="utf-8", errors='ignore')
    for line in current_file:
        if line != '\n': # standardize new lines
            compiled_news.write(line)
    compiled_news.write('\n\n') # add 2 newlines to separate each news

print(count)
        
```
